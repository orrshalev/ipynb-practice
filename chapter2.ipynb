```python
import numpy as np
P = np.array([0, 0, 0, 1, 2, 9])
s = np.array([P[5], P[4]])
n = len(s)
x_bar = np.mean(s)
biased_estimate = lambda s : np.sum(np.power((s - x_bar), 2)) / n
unbiased_estimate = lambda s : np.sum(np.power((s - x_bar), 2)) / (n - 1)
print(f"mean: {x_bar}")
print(f"biased estimate: {biased_estimate(s)}")
print(f"unbiased estimate: {unbiased_estimate(s)}")
```
